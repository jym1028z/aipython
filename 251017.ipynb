# 파이썬 함수

```python
# 함수 정의하기
def say_hello():
  print("안녕하세요! 만나서 반가워요.")
# 함수 호출(사용)하기
say_hello()
```

```python
def mood_check(score): # 1. 'score'라는 이름으로 '입력(재료)'을 받습니다.
# 2. if/elif/else 구문으로 입력받은 점수를 '처리(레시피)'합니다.
  if score >= 8:
# 3. 처리 결과에 따라 다른 '출력(요리)'을 return합니다.
    return "기분이 아주 좋아 보이네요! 즐거운 하루 되세요."
  elif score >= 5:
    return "그럭저럭 괜찮은 하루군요. 힘내세요."
  else:
    return "조금 우울하신가요? 따뜻한 차 한잔 어때요?"
user_score = int(input("오늘 기분 점수는? (0~10): "))
result_message = mood_check(user_score) # 함수가 반환한 결과가 result_message에 저장됩니다.
print(result_message)
```

함수 정의 후 호출. 정의할 때는 실행되는 것이 아님. 호출할 때 다시 올라가서 정의된 함수를 찾는 매커니즘

```python
def coffee_order(menu): # 'menu'라는 문자열을 입력으로 받습니다.
  if menu == "아메리카노":
    return "진한 에스프레소에 시원한 물 한 잔 추가했어요!"
  elif menu == "라떼":
    return "부드러운 우유 거품과 함께 나갑니다."
  elif menu == "바닐라라떼":
    return "달콤한 바닐라 향으로 기분 좋은 하루 시작하세요!"
  else:
    return "죄송해요, 해당 메뉴는 준비되지 않았어요."
drink = input("무슨 커피를 주문하시겠어요? ")
response = coffee_order(drink) # 사용자의 입력을 함수에 전달하고, 반환값을 받습니다.
print(response)
```

```python
def introduce(name, age):
  print(f"안녕하세요, 제 이름은 {name}이고, 나이는 {age}살이에요.")
# 함수 호출 시 두 개의 값을 순서에 맞게 전달합니다.
introduce("홍길동", 15)
```

```python
def greet(name="친구"):
  print(f"{name}님, 반가워요!")
greet() # 입력값이 없으면 기본값 '친구'가 사용됨 -> "친구님, 반가워요!"
greet("홍길동") # 입력값이 있으면 전달된 값 '홍길동'이 사용됨 -> "홍길동님, 반가워요!"
```

```python
def calculate_total_and_average(scores):
  total = sum(scores)
  average = total / len(scores)
  return total, average # 총점과 평균을 함께 반환
my_scores = [85, 90, 78, 92]
t, avg = calculate_total_and_average(my_scores) # 반환된 두 값을 각각 t와 avg 변수에 저장
print("총점:", t)
print("평균:", avg)
```

# 코딩실습 2

```python
def bmi(h,w):
  h=h*0.01
  bmi=w/h**2
  return round(bmi,2)

h=float(input("키(cm)를 입력하세요: "))
w=float(input("몸무게(kg)를 입력하세요: "))

print(bmi(h,w))
```

```python
# 실습 3-1: enumerate로 순위 매기기
scores = [95, 87, 92, 88]
print("성적 순위:")
for rank, score in enumerate(scores): # for rank, score in enumerate(scores, start=1): (정답)
  print(f" {rank}등: {score}점")
```

```python
# 실습 3-2: zip으로 학생 정보 결합
students = ["철수", "영희", "민수"]
scores = [85, 92, 78]
grades = ["B", "A", "C"]
print("\n학생 성적표:")
for name, score, grade in zip(students, scores, grades):
  print(f" {name}: {score}점 ({grade}학점)")
```

```python
# 실습 3-3: 딕셔너리 생성 (zip 활용)
keys = ["name", "age", "city"]
values = ["홍길동", 25, "부산"]
person_dict = dict(zip(keys, values))
print(f"\n딕셔너리: {person_dict}")
```

```python
# 실습 4-1: 섭씨를 화씨로 변환
celsius = [0, 10, 20, 30, 40]
fahrenheit=list(map(lambda x:x*1.8+32,celsius)) # fahrenheit = list(map(lambda c: c * 9/5 + 32, celsius)) (정답)
print(f"섭씨: {celsius}")
print(f"화씨: {fahrenheit}")
```

```python
print("\n" + "=" * 60)
print("5. filter - 원하는 것만 골라내기")
print("=" * 60)
```

```python
# 실습 5-1: 양수만 필터링
numbers = [-5, 3, -1, 8, 0, -3, 7]
positive=list(filter(lambda x:x>0,numbers)) # positive = list(filter(lambda x: x > 0, numbers)) (정답)
print(f"전체 숫자: {numbers}")
print(f"양수만: {positive}")
```

```python
# 실습 5-2: 길이가 5 이상인 단어만
words = ["hi", "hello", "python", "code", "programming"]
long_words=list(filter(lambda x:len(x)>=5,words)) # long_words = list(filter(lambda w: len(w) >= 5, words)) (정답)
print(f"\n전체 단어: {words}")
print(f"긴 단어: {long_words}")
```
