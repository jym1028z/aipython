3.1. 딕셔너리의 개념과 활용

```python
# 튜플(), 딕셔너리{}, 리스트[] 기호 구분하기(빈출)
# 튜플은 수정할 수 없음
```

```python
person = {
'name': '홍길동',
'age': 30,
'job': '개발자'
}
# 순회 - 하나씩 접근
```

```python
coffe={'아메리카노':2000,'라테':3000,'녹차':2500}
```

```python
coffe['아메리카노'] # 이 때는 [ ]이용
```

```python
for k in coffe:
  print(k) # 키 출력
  print(coffe[k]) # 딕셔너리에서 키를 지정하여 값 출력
```

```python
for k,v in coffe.items(): # items 함수를 이용해 키와 밸류를 동시에 순회. 가장 권장됨
  print(k,v)
```

전체 학생의 총점과 평균을 구하기, 가장 높은 점수를 받은 학생의 이름 뽑기

```python
score = {'홍길동':100,'이순신':90,'강감찬':80}

s=sum(score.values())
m=s/len(score)
M=max(score.values())
s,m,M
#print(f'총점:{s} / 평균: {m} / 가장 높은 점수를 받은 학생: {score[M]}')
```
