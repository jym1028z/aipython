```python
# Initialize a variable to store the current multiple of 3
multiple_of_3 = 0

# Use a loop to find the first four multiples
print("First four multiples of 3:")
for _ in range(5):
  multiple_of_3 += 3
  print(multiple_of_3)
```

### f-string

f-string 또는 형식화된 문자열 리터럴은 문자열 안에 표현식을 삽입하는 더 읽기 쉬운 방법을 제공합니다. 문자열 리터럴 앞에 'f' 또는 'F'를 붙여 만듭니다. 그런 다음 문자열 내의 중괄호 `{}` 안에 표현식을 넣을 수 있습니다.

```python
name = "Alice"
age = 30

# 변수를 삽입하기 위해 f-string 사용
greeting = f"안녕하세요, 제 이름은 {name}이고 {age}살입니다."
print(greeting)

# 중괄호 안에 표현식을 포함할 수도 있습니다.
print(f"2년 후에는 {age + 2}살이 될 것입니다.")
```

```python
a = 3.141592

# f-string을 사용하여 소수점 이하 두 자리까지 출력
print(f"소수점 두 자리까지: {a:.2f}")
```

원의 반지름을 입력받아 원의 넓이를 구하는 프로그램.


```python
r=float(input("원의 반지름을 입력해 주세요: "))
result=3.14*r**2
print(f'원의 넓이는 {result:.2f}입니다')
```
