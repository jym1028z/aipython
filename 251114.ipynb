```python
import pandas as pd
student_data = {
'name': ['Kim', 'Lee', 'Park'],
'test': [85, 92, 78],
'assign1': [88, 95, 80],
'assign2': [91, 89, 82]
}
df = pd.DataFrame(student_data)
display(df)
```

### 1. 데이터프레임 기본 정보 확인

```python
print('--- df.head() ---')
display(df.head())

print('\n--- df.tail() ---')
display(df.tail())

print('\n--- df.info() ---')
df.info()

print('\n--- df.describe() ---')
display(df.describe())

print('\n--- df.shape ---')
print(df.shape)

print('\n--- df.columns ---')
print(df.columns)

print('\n--- df.index ---')
print(df.index)

print('\n--- df.dtypes ---')
print(df.dtypes)
```

### 2. 특정 컬럼 정보 및 결측치 확인

```python
print('--- df["name"].value_counts() ---')
display(df['name'].value_counts())

print('\n--- df["test"].unique() ---')
print(df['test'].unique())

print('\n--- df.isnull().sum() (컬럼별 결측치 개수) ---')
display(df.isnull().sum())
```

### 3. 데이터 조작 (예시: 새로운 컬럼 추가 및 apply 사용)

```python
print('--- 새로운 컬럼 추가: total_score ---')
df['total_score'] = df['test'] + df['assign1'] + df['assign2']
display(df)

print('\n--- apply() 메소드를 사용하여 등급 부여 ---')
def get_grade(score):
    if score >= 260:
        return 'A'
    elif score >= 240:
        return 'B'
    else:
        return 'C'

df['grade'] = df['total_score'].apply(get_grade)
display(df)
```

### 4. 데이터 그룹화 (예시: groupby)

```python
print('--- grade별 평균 점수 ---')
display(df.groupby('grade')['total_score'].mean())
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 설치된 나눔 폰트 경로
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'

# FontManager에 폰트 추가 (이전 오류 해결 및 직접 등록)
fm.fontManager.addfont(font_path)
font_name = fm.FontProperties(fname=font_path).get_name()

# Matplotlib의 rcParams를 사용하여 폰트 설정
plt.rc('font', family=font_name)

# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

x = [1,2,3,4,5]
y=[2,4,6,8,10]

plt.title('기본 선형 그래프')
plt.bar(x,y)
plt.scatter(x,y)
plt.plot(x,y)
plt.show()

# 오 그냥 한 장에 그려짐
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

plt.figure(figsize=(10,7)) # 플롯 크기

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, s=100, alpha=0.8)
plt.title("iris 데이터 산점도", fontsize=16)
plt.xlabel("꽃받침 길이", fontsize=14)
plt.ylabel("꽃받침 너비", fontsize=14)
plt.legend(title='종류', fontsize=14)

plt.grid(True)

plt.show()
```
